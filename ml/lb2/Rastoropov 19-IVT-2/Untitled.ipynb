```python
def get_last_name(name):
    return name.split(' ')[0].strip()
company = cameras['Model'].apply(get_last_name)
cameras['Company'] = company
cameras.head()
```

```python
price3 = pd.DataFrame(cameras.groupby(['Release date'])['Price'].sum())
price3['count'] = price['Price']
```
